```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the first Gaussian distribution
mu1 = 590
sigma1 = 20
N = 1000  # Number of samples

# Generate random samples for the first Gaussian distribution
samples1 = np.random.normal(mu1, sigma1, N)

# Parameters for the second Gaussian distribution
mu2 = 500
sigma2 = 20

# Generate random samples for the second Gaussian distribution
samples2 = np.random.normal(mu2, sigma2, N)

# Create histograms for both sets of samples
nbins = 12
counts1, edges1 = np.histogram(samples1, bins=nbins)
counts2, edges2 = np.histogram(samples2, bins=nbins)

# Compute the x-axis for the histogram bars (midpoints of the bins)
xaxis1 = (edges1[1:] + edges1[:-1]) / 2
xaxis2 = (edges2[1:] + edges2[:-1]) / 2

# Normalize the histograms using trapezoidal integration
n_pdf1 = np.divide(counts1, np.trapz(counts1, xaxis1))
n_pdf2 = np.divide(counts2, np.trapz(counts2, xaxis2))

# Plot the first histogram as bars
plt.bar(xaxis1, n_pdf1, width=xaxis1[1] - xaxis1[0], alpha=0.5, label=f'Gaussian (mu={mu1}, sigma={sigma1})', color = "green" )
plt.plot(xaxis1, st.norm.pdf(xaxis1, mu1, sigma1), 'r-', linewidth=2)


# Plot the second histogram as bars
plt.bar(xaxis2, n_pdf2, width=xaxis2[1] - xaxis2[0], alpha=0.5, label=f'Gaussian (mu={mu2}, sigma={sigma2})', color = "grey")
plt.plot(xaxis2, st.norm.pdf(xaxis2, mu2, sigma2), 'r-', linewidth=2)



# Add labels and a legend
plt.title('Histograms of Two Gaussian Distributions')
plt.xlabel('# 1/4 turns per 30 min')
plt.ylabel('Fraction of Animals')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
